```python
import pandas as pd
import numpy as np
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
%matplotlib inline

filename='train.csv'
data=pd.read_csv(filename)

# female survival statistics

data_female=data.where(data['Sex']=='female')
female_num=data_female['Sex'].count()

data_female_survived=data.where(data_female['Survived']==1)
female_surv_num=data_female_survived['Survived'].count()

data_female_notsurvived=data.where(data_female['Survived']==0)
female_notsurv_num=data_female_notsurvived['Survived'].count()


## male survival statistics

data_male=data.where(data['Sex']=='male')
male_num=data_male['Sex'].count()

data_male_survived=data.where(data_male['Survived']==1)
male_surv_num=data_male_survived['Survived'].count()

data_male_notsurvived=data.where(data_male['Survived']==0)
male_notsurv_num=data_male_notsurvived['Survived'].count()


##data class

data_firstcls=data.where(data['PClass']==1)
data_firstcls_num=data_firstcls['PClass'].count()

data_firstcls_surv=data.where(data_firstcls['Survived']==1)
data_firstcls_notsurv=data.where(data_firstcls['Survived']==0)
data_firstcls_surv_num=data_firstcls_surv['Survived'].count()
data_firstcls_notsurv_num=data_firstcls_notsurv['Survived'].count()

data_firstcls_female=data.where(data_firstcls['Sex']=='female')
data_firstcls_female_survived=data.where(data_firstcls_female['Survived']==1)
data_firstcls_female_notsurvived=data.where(data_firstcls_female['Survived']==0)
data_firstcls_female_num=data_firstcls_female['Sex'].count()
data_firstcls_female_survived_num=data_firstcls_female_survived['Survived'].count()
data_firstcls_female_notsurvived_num=data_firstcls_female_notsurvived['Survived'].count()

data_firstcls_male=data.where(data_firstcls['Sex']=='male')
data_firstcls_male_survived=data.where(data_firstcls_male['Survived']==1)
data_firstcls_male_notsurvived=data.where(data_firstcls_male['Survived']==0)
data_firstcls_male_num=data_firstcls_male['Sex'].count()
data_firstcls_male_survived_num=data_firstcls_male_survived['Survived'].count()
data_firstcls_male_notsurvived_num=data_firstcls_male_notsurvived['Survived'].count()

data_secondcls=data.where(data['PClass']==2)
data_secondcls_num=data_secondcls['PClass'].count()

data_secondcls_surv=data.where(data_secondcls['Survived']==1)
data_secondcls_notsurv=data.where(data_secondcls['Survived']==0)
data_secondcls_surv_num=data_secondcls_surv['Survived'].count()
data_secondcls_notsurv_num=data_secondcls_notsurv['Survived'].count()

data_secondcls_female=data.where(data_secondcls['Sex']=='female')
data_secondcls_female_survived=data.where(data_secondcls_female['Survived']==1)
data_secondcls_female_notsurvived=data.where(data_secondcls_female['Survived']==0)
data_secondcls_female_num=data_secondcls_female['Sex'].count()
data_secondcls_female_survived_num=data_secondcls_female_survived['Survived'].count()
data_secondcls_female_notsurvived_num=data_secondcls_female_notsurvived['Survived'].count()

data_secondcls_male=data.where(data_secondcls['Sex']=='male')
data_secondcls_male_survived=data.where(data_secondcls_male['Survived']==1)
data_secondcls_male_notsurvived=data.where(data_secondcls_male['Survived']==0)
data_secondcls_male_num=data_secondcls_male['Sex'].count()
data_secondcls_male_survived_num=data_secondcls_male_survived['Survived'].count()
data_secondcls_male_notsurvived_num=data_secondcls_male_notsurvived['Survived'].count()

data_thirdcls=data.where(data['PClass']==3)
data_thirdcls_num=data_thirdcls['PClass'].count()

data_thirdcls_surv=data.where(data_thirdcls['Survived']==1)
data_thirdcls_notsurv=data.where(data_thirdcls['Survived']==0)
data_thirdcls_surv_num=data_thirdcls_surv['Survived'].count()
data_thirdcls_notsurv_num=data_thirdcls_notsurv['Survived'].count()

data_thirdcls_female=data.where(data_thirdcls['Sex']=='female')
data_thirdcls_female_survived=data.where(data_thirdcls_female['Survived']==1)
data_thirdcls_female_notsurvived=data.where(data_thirdcls_female['Survived']==0)
data_thirdcls_female_num=data_thirdcls_female['Sex'].count()
data_thirdcls_female_survived_num=data_thirdcls_female_survived['Survived'].count()
data_thirdcls_female_notsurvived_num=data_thirdcls_female_notsurvived['Survived'].count()

data_thirdcls_male=data.where(data_thirdcls['Sex']=='male')
data_thirdcls_male_survived=data.where(data_thirdcls_male['Survived']==1)
data_thirdcls_male_notsurvived=data.where(data_thirdcls_male['Survived']==0)
data_thirdcls_male_num=data_thirdcls_male['Sex'].count()
data_thirdcls_male_survived_num=data_thirdcls_male_survived['Survived'].count()
data_thirdcls_male_notsurvived_num=data_thirdcls_male_notsurvived['Survived'].count()


######data male/female stat clean

data_femalesurv_clean=data_female_survived.dropna()
data_malesurv_clean=data_male_survived.dropna()

data_femalenotsurv_clean=data_female_notsurvived.dropna()
data_malenotsurv_clean=data_male_notsurvived.dropna()


###data class clean

data_firstcls_surv_clean=data_firstcls_surv.dropna()
data_firstcls_notsurv_clean=data_firstcls_notsurv.dropna()

data_secondcls_surv_clean=data_secondcls_surv.dropna()
data_secondcls_notsurv_clean=data_secondcls_notsurv.dropna()

data_thirdcls_surv_clean=data_thirdcls_surv.dropna()
data_thirdcls_notsurv_clean=data_thirdcls_notsurv.dropna()
```

```python
#### graph age/class based on sex

graphvalue=[female_surv_num,female_notsurv_num,male_surv_num,male_notsurv_num]
colors=['blue','green','red','black']
labels=['females survived','females died','males survived','males died']

plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=2.0, hspace=0.9)

#plt.subplot(2,2,1)
sctr1=plt.scatter(data_femalenotsurv_clean['Age'],data_femalenotsurv_clean['PClass'],c='blue',marker='s',label='died')
sctr2=plt.scatter(data_femalesurv_clean['Age'],data_femalesurv_clean['PClass'],c='green',marker='d',label='survived')
plt.legend(bbox_to_anchor=(0.6,0.8), loc=3, borderaxespad=0.)
plt.title("Age Vs. Class : Female")
plt.xlabel("Age")
plt.ylabel("Class")
plt.savefig('sctr1.png')

plt.show()

```
